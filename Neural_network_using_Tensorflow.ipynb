Importing Libraries

```python
import tensorflow as tf
import numpy as np
from random import random
from sklearn.model_selection import train_test_split
```

Plan:

```python
# build model
# compile model
# train model
# evaluate model
# make predictions
```

Main Code:

```python
#generating dataset:
def generate_dataset(num_samples, test_size):
  x = np.array([[random()/2 for _ in range(2)] for _ in range(num_samples)])
  y = np.array([[i[0] + i[1]] for i in x])
  x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=test_size)
  return x_train, x_test, y_train, y_test


#spliting data
x_train, x_test, y_train, y_test = generate_dataset(5000,0.3)

#Building model
#Our model is will have 2 input nodes, followed by 5 nodes and 1 output node
#2 -> 5 -> 1

model = tf.keras.Sequential([
            tf.keras.layers.Dense(5, input_dim = 2, activation='sigmoid'), #hidden layer
            tf.keras.layers.Dense(1, activation='sigmoid') #output layer
            ])

#optimiser will Stochastic Gradient Descent
optimiser =  tf.keras.optimizers.SGD(learning_rate = 0.1)

#Compiling the model
#The optimiser will be SGD and loss function will be Mean Square error
model.compile(optimizer=optimiser, loss='MSE')

#Training the model
model.fit(x_train, y_train, epochs = 100)

#Evaluating the model
print("\n Model Evaluation: ")
model.evaluate(x_test,y_test, verbose=1)

#Making Predictions
data = np.array([[0.1,0.2],[0.2,0.2]])
predictions = model.predict(data)

#Printing the predictions
print("\n Some predictions: ")

for d,p in zip(data,predictions):
  print("{} + {} = {}".format(d[0],d[1],p[0]))

```
